b) write a function dup to find all duplicates in the list

```python
def dups(numlist):
  temp=[]
  for i in numlist:
    cnt=numlist.count(i)
    if(cnt>1 and (i not in temp)):
      temp.append(i)
      print("duplicates in the list",temp)
      1st=[1,2,6,9,1,3,5,2,1]
      print("Given elements in the list",1st)
      dups(1st)
```
